```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    "health-lifestyle-data-analysis/data/sleep_health_lifestyle_dataset.csv"
)

df.head()
```

```python
df.info()

```

```python
df.isnull().sum()
```

```python
df.describe()
```

df.columns


```python
df.groupby("Stress Level (scale: 1-10)")["Sleep Duration (hours)"].mean().plot(kind="bar")
plt.title("Average Sleep Duration by Stress Level")
plt.xlabel("Stress Level")
plt.ylabel("Sleep Duration (hours)")
plt.show()
```

```python
df.groupby("Physical Activity Level (minutes/day)")[
    "Quality of Sleep (scale: 1-10)"
].mean().plot(kind="line", marker="o")

plt.title("Physical Activity vs Quality of Sleep")
plt.xlabel("Physical Activity Level (minutes/day)")
plt.ylabel("Quality of Sleep")
plt.show()
```

```python
df.groupby("BMI Category")["Sleep Duration (hours)"].mean().plot(kind="bar")
plt.title("Average Sleep Duration by BMI Category")
plt.ylabel("Sleep Duration (hours)")
plt.show()
```
